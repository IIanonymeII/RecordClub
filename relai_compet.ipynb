```python
import pandas as pd
```

```python
df = pd.read_csv("final_relai.csv")
df.head()


# Convert the 'Time' column to datetime format
df['Time'] = pd.to_datetime(df['Time'], format='%Y-%m-%d %H:%M:%S.%f', errors='coerce')

df.head()
```

```python

df_homme = df[df['Sex'] == "H"]
df_femme = df[df['Sex'] == "F"]
df_mixte = df[df['Sex'] == "A"]


df_femme.head()
```

```python

df_homme_25 = df_homme[df_homme['Bassin'] == 25]
df_homme_50 = df_homme[df_homme['Bassin'] == 50]

df_femme_25 = df_femme[df_femme['Bassin'] == 25]
df_femme_50 = df_femme[df_femme['Bassin'] == 50]

df_mixte_25 = df_mixte[df_mixte['Bassin'] == 25]
df_mixte_50 = df_mixte[df_mixte['Bassin'] == 50]
```

```python
df_femme_50.info()
```

```python
list_femme_by_cat_25 = {}
list_femme_by_cat_50 = {}

list_homme_by_cat_25 = {}
list_homme_by_cat_50 = {}


list_mixte_by_cat_25 = {}
list_mixte_by_cat_50 = {}

for limit_age in range(12,19,1):
    min_time_indices_f_25 = df_femme_25[df_femme_25["Max_age"]<=limit_age].groupby('Type')['Time'].idxmin()
    min_time_indices_f_50 = df_femme_50[df_femme_50["Max_age"]<=limit_age].groupby('Type')['Time'].idxmin()

    min_time_indices_h_25 = df_homme_25[df_homme_25["Max_age"]<=limit_age].groupby('Type')['Time'].idxmin()
    min_time_indices_h_50 = df_homme_50[df_homme_50["Max_age"]<=limit_age].groupby('Type')['Time'].idxmin()

    list_femme_by_cat_25[limit_age] = df.loc[min_time_indices_f_25]
    list_femme_by_cat_50[limit_age] = df.loc[min_time_indices_f_50]

    list_homme_by_cat_25[limit_age] = df.loc[min_time_indices_h_25]
    list_homme_by_cat_50[limit_age] = df.loc[min_time_indices_h_50]


min_time_indices_f_25 = df_femme_25.groupby('Type')['Time'].idxmin()
min_time_indices_f_50 = df_femme_50.groupby('Type')['Time'].idxmin()

min_time_indices_h_25 = df_homme_25.groupby('Type')['Time'].idxmin()
min_time_indices_h_50 = df_homme_50.groupby('Type')['Time'].idxmin()

min_time_indices_a_25 = df_mixte_25.groupby('Type')['Time'].idxmin()
min_time_indices_a_50 = df_mixte_50.groupby('Type')['Time'].idxmin()

list_femme_by_cat_25["all"] = df.loc[min_time_indices_f_25]
list_femme_by_cat_50["all"] = df.loc[min_time_indices_f_50]

list_homme_by_cat_25["all"] = df.loc[min_time_indices_h_25]
list_homme_by_cat_50["all"] = df.loc[min_time_indices_h_50]

list_mixte_by_cat_25["all"] = df.loc[min_time_indices_a_25]
list_mixte_by_cat_50["all"] = df.loc[min_time_indices_a_50]
```

```python
list_homme_by_cat_50["all"]
```

```python
need_column = ['Full Name', 'Bassin','Type', 'Time', 'Age', 'Where', 'Date']
df_record = list_femme_by_cat_25["all"][need_column]

# Change the format to %M:%S.%f
df_record['Time'] = df_record['Time'].dt.strftime('%M:%S.%f').str[:8]

df_record
```
